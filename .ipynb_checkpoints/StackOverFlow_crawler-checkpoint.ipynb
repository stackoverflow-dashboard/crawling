```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver import ActionChains
from time import sleep
from time import time
import pandas
from bs4 import BeautifulSoup
```

```python
def SOFCrawling():
    # driver 생성 및 get
    options = webdriver.ChromeOptions()
    user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"
    options.add_argument('user-agent=' + user_agent)
    driver = webdriver.Chrome(options=options)
    driver.get("https://stackoverflow.com/questions?tab=newest&page=1") # 첫 페이지
    html = driver.page_source
    soup = BeautifulSoup(html,'html')
    
    # 50개 단위로 설정
    button = driver.find_element(By.XPATH,'//*[@id="mainbar"]/div[4]/a[3]')
    ActionChains(driver).click(button).perform()
    
    sleep(1)
    
    for i in range(1,10001):
        # 쿠키 설정 팝업 - 모든 쿠키 허용 - 으로 설정
        try:
            elem = driver.find_element(By.ID, "onetrust-accept-btn-handler")
            elem.click()
        except:
            pass
        
        # 크롤링 실시
        infoCrawling(driver, dic)
        
        # 다음 페이지로
        button = driver.find_element(By.XPATH, '//a[@class="s-pagination--item js-pagination-item" and @rel="next"]')
        ActionChains(driver).click(button).perform()
        sleep(0.2)
        
    # 크롤링 종료
    driver.quit()
    
    
def infoCrawling(driver, data) :
    for elem in driver.find_elements(By.CLASS_NAME, "s-post-summary.js-post-summary"):
        
        # question_id
        question_id = elem.get_attribute("data-post-id")
        dic['question-id'].append(question_id)
        
        # title
        title = driver.find_element(By.XPATH, f'//*[@id="question-summary-{question_id}"]/div[2]/h3').text
        dic['title'].append(title)
        
        # date
        tmp_date = driver.find_element(By.XPATH, f'//*[@id="question-summary-{question_id}"]/div[2]/div[2]/div[2]/time/span')
        date = tmp_date.get_attribute("title")
        dic['date'].append(date)
        
        # answer_cnt, view_cnt
        answer_cnt = driver.find_element(By.XPATH, f'//*[@id="question-summary-{question_id}"]/div[1]/div[2]/span[1]').text
        dic['answer_cnt'].append(answer_cnt)
        view_cnt = driver.find_element(By.XPATH, f'//*[@id="question-summary-{question_id}"]/div[1]/div[3]/span[1]').text
        dic['view_cnt'].append(view_cnt)
        
        # tag
        tags = driver.find_element(By.XPATH, f'//*[@id="question-summary-{question_id}"]/div[2]/div[2]/div[1]/ul')
        tmp_tags = []
        for tag in tags.find_elements(By.TAG_NAME, "li"):
            tmp_tags.append(tag.text)
        dic['tags'].append(tmp_tags)
                                    
if __name__ == '__main__' :
     # 받을 데이터 딕셔너리 준비
    dic = {'question-id':[],'title':[],'date':[],'tags':[],'answer_cnt':[],'view_cnt':[]}
                                    
    # 실행
    SOFCrawling()
    
    #csv 파일로 저장
    df = pandas.DataFrame(dic, columns=["question-id", "title", "date", "tags", "answer_cnt", "view_cnt"])
    print(df.head(10))

    df.to_csv(f"C:\\Users\\yuns1\\DevCourse\\2차 프로젝트\\raw_data\\SOF_info.csv", encoding='utf-8-sig', index=False)
                                    
```
